Transformamos la fecha de nacimiento a edad, contando con el día **25/09/2023**

```python
# Create a function
from datetime import datetime
def calculate_age(df, dob_column_name, target_date_str):
    df[dob_column_name] = pd.to_datetime(df[dob_column_name])
    target_date = pd.to_datetime(target_date_str)
    df['age'] = (target_date - df[dob_column_name]).dt.days // 365
    return df

full_df = calculate_age(full_df, 'dob', '2023-09-25')
#Dropeamos la fecha de nacimiento
full_df = full_df.drop(['dob'], axis = 1)
```

```python
def check_holidays(df1, df2, date_column_name, result_column_name):
    # Assuming df1 and df2 have a 'date_column_name' column containing dates
    df1[date_column_name] = pd.to_datetime(df1[date_column_name]).dt.date
    df2[date_column_name] = pd.to_datetime(df2[date_column_name]).dt.date

    # Merge dataframes on the specified date column
    merged_df = pd.merge(df1, df2, on=date_column_name, how='left')

    # Create a new column in df1 to store the check result
    df1[result_column_name] = df1[date_column_name].isin(df2[date_column_name])

    return df1
```
